## 🧮 The Mathematical Magic Formula

For any binomial experiment, the probability of getting exactly $k$ successes in $n$ trials is:

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

Where:
- $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ is the **"n choose k"** combination
- $p$ is the probability of success
- $(1-p)$ is the probability of failure
- $k$ is the number of successes we want

## 🧮 Act 2: The Mathematical Magic Formula! ✨

Now let's peek behind the curtain and discover the **mathematical magic** that makes our binomial distribution work! 

### 🎭 The Secret Formula of Our Circus:

For any binomial experiment, the probability of getting exactly $k$ successes in $n$ trials is:

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

### 🎪 Breaking Down the Magic:

1. **$\binom{n}{k} = \frac{n!}{k!(n-k)!}$** - The **"n choose k"** combination
   - 🎯 **What it does**: Counts how many different ways we can get exactly $k$ successes
   - 🎪 **Think of it as**: How many different sequences of heads and tails give us exactly $k$ heads?

2. **$p^k$** - The probability of $k$ successes
   - 🎯 **What it does**: Probability of getting $k$ heads
   - 🎪 **Think of it as**: The chance of getting heads $k$ times in a row

3. **$(1-p)^{n-k}$** - The probability of $n-k$ failures
   - 🎯 **What it does**: Probability of getting $n-k$ tails
   - 🎪 **Think of it as**: The chance of getting tails the remaining times

4. **The Product** - Total probability
   - 🎯 **What it does**: Combines all the possibilities
   - 🎪 **Think of it as**: The total magic of getting exactly $k$ heads!

### 🎲 Let's Try the Magic with Our Coin! 🪙

For our magical coin show with $n=10$, $p=0.5$, $k=4$:

- **$\binom{10}{4} = \frac{10!}{4!6!} = 210$** - There are 210 different ways to get exactly 4 heads!
- **$P(X=4) = 210 \times (0.5)^4 \times (0.5)^6 = 210 \times 0.5^{10} \approx 0.205$**
- **🎯 This means**: There's about a 20.5% chance of getting exactly 4 heads in 10 flips!

### 🎪 The Magic of Combinations! 🎭

The combination $\binom{n}{k}$ is like asking: *"How many different ways can I arrange $k$ successes among $n$ trials?"*

For example, with 4 heads in 10 flips:
- HHHHTTTTTT (heads first)
- THHHHTTTT (heads in the middle)
- TTTTTHHHH (heads last)
- And 207 other ways! 🎪

## 🎪 Let's Start with the Pizza Delivery Challenge! 🍕

This is a perfect example of how binomial distributions help us make real-world decisions!

```python
# 🍕 The Pizza Delivery Challenge! 🍕
print("🍕" + "="*70 + "🍕")
print("🍕 THE PIZZA DELIVERY CHALLENGE! 🍕")
print("🍕" + "="*70 + "🍕")
print("🎯 You're a pizza delivery driver with an 80% on-time delivery rate!")
print("🎪 You have 15 deliveries today. Let's see how you'll do! 🎪")

# 🍕 Pizza delivery parameters
n_deliveries = 15  # Number of deliveries today
p_success = 0.8   # 80% success rate for on-time delivery
target_deliveries = 12  # We want at least 12 on-time deliveries

print(f"🍕 Today's Challenge:")
print(f"   - Total deliveries: {n_deliveries}")
print(f"   - Success rate: {p_success*100}%")
print(f"   - Target: At least {target_deliveries} on-time deliveries")
print(f"   - Expected on-time deliveries: {n_deliveries * p_success}")

# 🧮 Calculate the probability using binomial distribution
# P(X >= 12) = 1 - P(X <= 11)
prob_at_least_12 = 1 - binom.cdf(target_deliveries - 1, n_deliveries, p_success)
prob_exactly_12 = binom.pmf(target_deliveries, n_deliveries, p_success)

print(f"\n🎯 Probability Analysis:")
print(f"   - Probability of exactly {target_deliveries} on-time deliveries: {prob_exactly_12:.4f} ({prob_exactly_12*100:.2f}%)")
print(f"   - Probability of at least {target_deliveries} on-time deliveries: {prob_at_least_12:.4f} ({prob_at_least_12*100:.2f}%)")

# 🎪 Let's simulate this scenario multiple times
N_simulations = 10000
simulated_results = np.random.binomial(n_deliveries, p_success, N_simulations)

# 🎯 Count how many times we achieved our target
successful_days = np.sum(simulated_results >= target_deliveries)
empirical_prob = successful_days / N_simulations

print(f"\n🎪 Simulation Results ({N_simulations:,} days):")
print(f"   - Days with at least {target_deliveries} on-time deliveries: {successful_days:,}")
print(f"   - Empirical probability: {empirical_prob:.4f} ({empirical_prob*100:.2f}%)")
print(f"   - Difference from theoretical: {abs(empirical_prob - prob_at_least_12):.4f}")

# 🎨 Create a spectacular visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('🍕 The Pizza Delivery Challenge! 🍕', fontsize=24, fontweight='bold', y=0.98)

# 🎨 Plot 1: Distribution of on-time deliveries
axes[0, 0].hist(simulated_results, bins=range(0, n_deliveries+2), color=colors[0], alpha=0.8, 
                edgecolor='black', linewidth=2)
axes[0, 0].axvline(target_deliveries, color='red', linestyle='--', linewidth=3, 
                   label=f'🎯 Target: {target_deliveries} deliveries')
axes[0, 0].axvline(n_deliveries * p_success, color='green', linestyle='--', linewidth=3, 
                   label=f'📊 Expected: {n_deliveries * p_success:.1f} deliveries')
axes[0, 0].set_title('🍕 Distribution of On-Time Deliveries 🍕', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Number of On-Time Deliveries', fontsize=14)
axes[0, 0].set_ylabel('Frequency', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# 🎨 Plot 2: Probability of different outcomes
outcomes = list(range(n_deliveries + 1))
probabilities = [binom.pmf(x, n_deliveries, p_success) for x in outcomes]

axes[0, 1].bar(outcomes, probabilities, color=colors[1], alpha=0.8, edgecolor='black', linewidth=2)
axes[0, 1].axvline(target_deliveries, color='red', linestyle='--', linewidth=3, 
                   label=f'🎯 Target: {target_deliveries}')
axes[0, 1].set_title('🎲 Probability of Each Outcome 🎲', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Number of On-Time Deliveries', fontsize=14)
axes[0, 1].set_ylabel('Probability', fontsize=14)
axes[0, 1].legend(fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# 🎨 Plot 3: Cumulative probability
cumulative_probs = [binom.cdf(x, n_deliveries, p_success) for x in outcomes]
axes[1, 0].plot(outcomes, cumulative_probs, 'o-', color=colors[2], linewidth=3, markersize=8)
axes[1, 0].axvline(target_deliveries, color='red', linestyle='--', linewidth=3, 
                   label=f'🎯 Target: {target_deliveries}')
axes[1, 0].set_title('📈 Cumulative Probability 📈', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Number of On-Time Deliveries', fontsize=14)
axes[1, 0].set_ylabel('Cumulative Probability', fontsize=14)
axes[1, 0].legend(fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# 🎨 Plot 4: Success vs Failure days
success_days = np.sum(simulated_results >= target_deliveries)
failure_days = N_simulations - success_days

labels = ['🎯 Success Days', '❌ Failure Days']
sizes = [success_days, failure_days]
colors_pie = [colors[3], colors[4]]

wedges, texts, autotexts = axes[1, 1].pie(sizes, labels=labels, colors=colors_pie, autopct='%1.1f%%', 
                                         startangle=90, textprops={'fontsize': 12})
axes[1, 1].set_title('🍕 Success vs Failure Days 🍕', fontsize=16, fontweight='bold')

# 🎨 Make the pie chart more beautiful
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(14)

plt.tight_layout()
plt.show()

# 🎪 Print the final results
print("\n🍕" + "="*70 + "🍕")
print("🍕 FINAL PIZZA DELIVERY CHALLENGE RESULTS! 🍕")
print("🍕" + "="*70 + "🍕")
print(f"🎯 Theoretical probability of success: {prob_at_least_12:.4f} ({prob_at_least_12*100:.2f}%)")
print(f"🎪 Empirical probability of success: {empirical_prob:.4f} ({empirical_prob*100:.2f}%)")
print(f"📊 Expected on-time deliveries: {n_deliveries * p_success:.1f}")
print(f"📈 Standard deviation: {np.sqrt(n_deliveries * p_success * (1 - p_success)):.2f}")
print(f"🎪 You have a {prob_at_least_12*100:.1f}% chance of meeting your target! 🎪")
print("🍕" + "="*70 + "🍕")
```

# Problem 2:

Now let's consider a total of $500000$ groups of $100$ people in size and the probability of voting for candidate $A$ to be $[25\%, 50\%, 75\%]$. Plot the probability of candidate $A$ to obtain $[0, 1, 2, 3, ..., 100]$ votes in each group. In other words plot the probability density function and compare them to the theoretical results. Describe your observations. 

```python
N_groups = 50000
n_group = 100

p_A_list = [0.25, 0.5, 0.75]

prob = []
prob_theory = []

for p_A in p_A_list:

    probability = []                 # define an array that will hold the calculated probability based on the flips
    theoretical_probability = []     # define an array that will hold the theoretical probability

    # result = []
    # for i in range(N):
    #     trial = np.random.rand(n_group) <= p_A # returns a list of n_group boolean function (True = H, False = T)
    #     result.append( np.sum(trial)) # computes the total number of supporters for candidate A for each trial
    # print(result)
    result = np.random.binomial(n_group, p_A, size = N_groups)

    for x in range(n_group+1):

        cases = 0 

        for item in result:
            if item == x:
                cases += 1

        probability_cases = cases / N_groups
        probability.append(probability_cases)

        theoretical_result = ( fact(n_group) / ( fact(x) * fact(n_group-x) ) ) * (p_A**x) * (1-p_A)**(n_group-x)
        theoretical_probability.append(theoretical_result)

    prob.append(probability)
    prob_theory.append(theoretical_probability)
```

```python
people = np.arange(n_group+1)

plt.figure(figsize = (15,7))

for index in range(len(p_A_list)):
    plt.plot(people, prob[index], linewidth = 4, label = f'numerical, $p_A$ = {p_A_list[index]}')
    plt.plot(people, prob_theory[index], "-.", color  = 'black', linewidth = 2)
plt.plot(people, prob_theory[index], "-.", color  = 'black', linewidth = 2 ,label = 'theoretical') # for legend purposes (not elegant, but does the job)

plt.legend(fontsize = 12, loc = 'upper right')

plt.title('Binomial Distribution\n Group Size: $n = $'+str(n_group)+' people', fontsize = 30)

plt.xlabel ('Number of votes, $A$', fontsize = 25)
plt.ylabel ('Probability, $p(A)$', fontsize = 25)

plt.show()
```
