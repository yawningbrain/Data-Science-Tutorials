## 🔍 Understanding Hypothesis Testing: The Detective's Method 🔍

### 🎯 What is Hypothesis Testing?
**Hypothesis testing** is a systematic way to make decisions about population parameters based on sample data. It's like being a detective who must decide whether evidence supports a claim or if it could have happened by chance.

### 🎪 The Two Competing Hypotheses:

#### 🛡️ **Null Hypothesis (H₀)**: "Innocent Until Proven Guilty"
- **Definition**: A statement of **no difference** or **no effect**
- **Purpose**: The default assumption we test against
- **Examples**: 
  - "The new drug has no effect on blood pressure"
  - "There's no difference in sugar consumption between 1961 and 2013"
  - "The coin is fair (50% heads, 50% tails)"

#### 🎯 **Alternative Hypothesis (Hₐ)**: "The Claim We're Investigating"
- **Definition**: A claim that **contradicts** the null hypothesis
- **Purpose**: What we conclude if we reject H₀
- **Examples**:
  - "The new drug reduces blood pressure"
  - "Sugar consumption has changed since 1961"
  - "The coin is biased"

### 🧮 The Decision Process:
1. **🔍 Collect Evidence**: Gather sample data
2. **🧮 Calculate Test Statistic**: Measure how extreme our sample is
3. **📊 Find P-value**: Probability of seeing this result by chance
4. **🎯 Make Decision**: Compare p-value to significance level (α)

### 🎪 Decision Options:
- **"Reject H₀"**: Evidence strongly supports the alternative hypothesis
- **"Fail to Reject H₀"**: Evidence is insufficient to reject the null hypothesis
- **Note**: We never "accept" H₀ - we either reject it or fail to reject it!

### 🎯 Significance Level (α): Our Threshold for "Reasonable Doubt"
- **Definition**: The probability of rejecting H₀ when it's actually true
- **Common Values**: α = 0.05 (5%), α = 0.01 (1%), α = 0.10 (10%)
- **Interpretation**: "We're willing to be wrong α% of the time"

### 🚨 The Two Types of Errors:

#### 🚨 **Type I Error (False Positive)**: "Crying Wolf"
- **Definition**: Rejecting a **true** null hypothesis
- **Probability**: α (our significance level)
- **Example**: Concluding a drug works when it doesn't
- **Risk**: Wasting resources on ineffective treatments

#### 🚨 **Type II Error (False Negative)**: "Missing the Signal"
- **Definition**: Failing to reject a **false** null hypothesis
- **Probability**: β (depends on sample size and effect size)
- **Example**: Missing that a drug actually works
- **Risk**: Missing opportunities for effective treatments

### 🎨 The Error Trade-off:
- **Decreasing α**: Reduces Type I errors but increases Type II errors
- **Increasing Sample Size**: Reduces both Type I and Type II errors
- **Larger Effect Size**: Reduces Type II errors

- **🎓 Education**: Evaluating teaching methods

```python
# 🌟 Welcome to the Statistical Detective Agency! Let's set up our tools! 🌟
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad 
from scipy.stats import norm, sem
import random
import warnings
warnings.filterwarnings('ignore')

# 🎨 Let's make our plots beautiful and detective-worthy! ✨
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3

# 🎯 Set up our detective color palette
detective_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F']

print("🌟 Welcome to the Statistical Detective Agency! 🌟")
print("✨ All detective tools loaded successfully! ✨")
print("🔍 Ready to investigate statistical mysteries! 🔍")
print("🎪 Let's solve this hypothesis testing case! 🎪")
```

```python
# 🧮 Creating Our Detective's Hypothesis Testing Function! 🧮
def standard_normal(x):
    """
    🎯 Standard normal probability density function
    
    Parameters:
    -----------
    x : float
        Value to evaluate
        
    Returns:
    --------
    float : PDF value at x
    """
    f_x = (1 / np.sqrt(2*np.pi)) * np.exp(-(x**2) / 2)
    return f_x

def hypothesis_test_detective(sample, sample_size, mu_baseline, sigma_baseline, alpha, year):
    """
    🔍 Statistical Detective's Hypothesis Testing Function
    
    Parameters:
    -----------
    sample : array-like
        Sample data from the year being tested
    sample_size : int
        Size of the sample
    mu_baseline : float
        Population mean from baseline year (1961)
    sigma_baseline : float
        Population standard deviation from baseline year
    alpha : float
        Significance level (e.g., 0.05 for 5%)
    year : int
        Year being tested
        
    Returns:
    --------
    dict : Results of the hypothesis test
    """
    
    # 🎯 Calculate sample statistics
    x_bar = np.mean(sample)
    z_score = (x_bar - mu_baseline) / (sigma_baseline / np.sqrt(sample_size))
    
    # 🎪 Display hypotheses
    print(f"🔍" + "="*60 + "🔍")
    print(f"🔍 HYPOTHESIS TEST FOR YEAR {year} 🔍")
    print(f"🔍" + "="*60 + "🔍")
    print(f"🛡️ H₀ (Null Hypothesis): Global sugar consumption has NOT changed since 1961")
    print(f"   📊 Baseline mean (1961): {mu_baseline:.2f} calories/day")
    print(f"🎯 Hₐ (Alternative Hypothesis): Global sugar consumption HAS changed since 1961")
    print(f"   📊 Sample mean ({year}): {x_bar:.2f} calories/day")
    print(f"   📈 Difference: {x_bar - mu_baseline:.2f} calories/day")
    print(f"   🧮 Z-score: {z_score:.3f}")
    
    # 🧮 Calculate p-value (two-tailed test)
    if z_score >= 0:
        # Right tail
        p_val = 2 * (1 - norm.cdf(z_score))
    else:
        # Left tail
        p_val = 2 * norm.cdf(z_score)
    
    print(f"\n📊 Statistical Analysis:")
    print(f"   🎯 Sample size: {sample_size}")
    print(f"   📈 P-value: {p_val:.4f} ({p_val*100:.2f}%)")
    print(f"   🎪 Significance level (α): {alpha:.3f} ({alpha*100:.1f}%)")
    
    # 🎯 Make decision
    if p_val < alpha:
        decision = "REJECT H₀"
        error_type = "Type I Error"
        error_description = "We conclude consumption has changed when it actually hasn't"
        print(f"\n🚨 DECISION: {decision}")
        print(f"   ✅ P-value ({p_val:.4f}) < α ({alpha:.3f})")
        print(f"   🎯 Conclusion: Evidence suggests sugar consumption HAS changed since 1961")
    else:
        decision = "FAIL TO REJECT H₀"
        error_type = "Type II Error"
        error_description = "We conclude consumption hasn't changed when it actually has"
        print(f"\n🚨 DECISION: {decision}")
        print(f"   ✅ P-value ({p_val:.4f}) ≥ α ({alpha:.3f})")
        print(f"   🎯 Conclusion: Insufficient evidence to claim sugar consumption has changed")
    
    print(f"\n🚨 Potential Error Analysis:")
    print(f"   🎪 Error Type: {error_type}")
    print(f"   📊 Description: {error_description}")
    print(f"   🎯 Probability: {alpha:.1%} (Type I) or depends on effect size (Type II)")
    
    # 🎪 Return results
    results = {
        'year': year,
        'sample_mean': x_bar,
        'baseline_mean': mu_baseline,
        'difference': x_bar - mu_baseline,
        'z_score': z_score,
        'p_value': p_val,
        'alpha': alpha,
        'decision': decision,
        'error_type': error_type,
        'significant': p_val < alpha
    }
    
    return results

print("✅ Detective's hypothesis testing function created successfully!")
print("🔍 Ready to investigate statistical mysteries! 🔍")
```

```python
# 🌍 Loading Our Global Nutrition Dataset! 🌍
print("🌍 Loading global dietary consumption data...")
print("🎪 This dataset contains nutrition data from countries worldwide! 🎪")

# 📊 Load the dataset
df = pd.read_csv('./Datasets/dietary-compositions-by-commodity-group.csv')

print("✅ Dataset loaded successfully!")
print(f"📊 Dataset shape: {df.shape}")
print(f"🌍 Columns available: {list(df.columns)}")
print(f"📅 Years in dataset: {sorted(df['Year'].unique())}")
print(f"🏛️ Number of countries: {df['Entity'].nunique()}")

# 🎨 Let's take a peek at our data
print("\n🎨 First few rows of our dataset:")
display(df.head())

print("\n🎨 Dataset info:")
print(df.info())

print("\n🎨 Basic statistics:")
display(df.describe())

print("🌟 Our global nutrition dataset is ready for investigation! 🌟")
```

```python
# 🎯 Step 1: Setting Up Our Investigation Parameters! 🎯
print("🎯 Step 1: Setting up our statistical investigation!")
print("🔍 We'll investigate sugar consumption changes over time! 🔍")

# 🍭 Choose our food category
category = 'Sugar (FAO (2017)) (kilocalories per person per day)'
year_baseline = 1961

print(f"\n📊 Investigation Setup:")
print(f"   🍭 Food Category: Sugar consumption")
print(f"   📅 Baseline Year: {year_baseline}")
print(f"   🎯 Population Parameter: Global average sugar consumption in {year_baseline}")

# 📊 Extract baseline data (1961)
df_baseline = df[df.Year == year_baseline]
baseline_data = df_baseline[category].dropna()

mu_baseline = np.mean(baseline_data)
sigma_baseline = np.std(baseline_data)

print(f"\n📊 Baseline Statistics ({year_baseline}):")
print(f"   🌍 Number of countries: {len(baseline_data)}")
print(f"   📈 Mean consumption: {mu_baseline:.2f} calories/day")
print(f"   📊 Standard deviation: {sigma_baseline:.2f} calories/day")
print(f"   🎯 Min consumption: {np.min(baseline_data):.2f} calories/day")
print(f"   🎯 Max consumption: {np.max(baseline_data):.2f} calories/day")

# 🎪 Set up investigation parameters
years_to_test = np.arange(1962, 2013, 5)
alpha = 0.05  # 5% significance level
sample_size = 50

print(f"\n🎪 Investigation Parameters:")
print(f"   📅 Years to test: {list(years_to_test)}")
print(f"   🎯 Significance level (α): {alpha:.3f} ({alpha*100:.1f}%)")
print(f"   📊 Sample size: {sample_size} countries per test")

# 🎨 Visualize baseline distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('🍭 Baseline Sugar Consumption Distribution (1961) 🍭', fontsize=16, fontweight='bold')

# 🎨 Plot 1: Histogram
axes[0].hist(baseline_data, bins=30, color=detective_colors[0], alpha=0.8, edgecolor='black')
axes[0].axvline(mu_baseline, color='red', linestyle='--', linewidth=3, label=f'Mean: {mu_baseline:.1f}')
axes[0].set_title('📊 Distribution of Sugar Consumption', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sugar Consumption (calories/day)', fontsize=12)
axes[0].set_ylabel('Number of Countries', fontsize=12)
axes[0].legend(fontsize=12)
axes[0].grid(True, alpha=0.3)

# 🎨 Plot 2: Box plot
bp = axes[1].boxplot(baseline_data, vert=True, patch_artist=True,
                    boxprops=dict(facecolor=detective_colors[1], alpha=0.8, linewidth=2),
                    medianprops=dict(color='red', linewidth=3),
                    whiskerprops=dict(linewidth=2),
                    capprops=dict(linewidth=2))
axes[1].set_title('📦 Box Plot of Sugar Consumption', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Sugar Consumption (calories/day)', fontsize=12)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("🌟 Baseline analysis complete! Ready to test for changes over time! 🌟")
```

```python
# 🎪 Step 2: Running Our Hypothesis Tests! 🎪
print("🎪 Step 2: Running hypothesis tests for each year!")
print("🔍 Let's see which years show significant changes in sugar consumption! 🔍")

# 🎯 Store results for analysis
all_results = []

# 🎪 Test each year
for year in years_to_test:
    print(f"\n{'='*80}")
    print(f"🔍 TESTING YEAR {year} 🔍")
    print(f"{'='*80}")
    
    # 📊 Get data for this year
    year_data = df[df.Year == year][category].dropna()
    
    if len(year_data) < sample_size:
        print(f"⚠️ Warning: Only {len(year_data)} countries available for {year}")
        print(f"   Skipping this year due to insufficient data.")
        continue
    
    # 🎲 Take a random sample
    sample = random.sample(list(year_data), sample_size)
    
    # 🧮 Run hypothesis test
    result = hypothesis_test_detective(
        sample=sample,
        sample_size=sample_size,
        mu_baseline=mu_baseline,
        sigma_baseline=sigma_baseline,
        alpha=alpha,
        year=year
    )
    
    # 🎯 Store results
    all_results.append(result)
    
    print(f"\n🎯 Year {year} Summary:")
    print(f"   📊 Sample mean: {result['sample_mean']:.2f} calories/day")
    print(f"   📈 Difference from baseline: {result['difference']:.2f} calories/day")
    print(f"   🧮 Z-score: {result['z_score']:.3f}")
    print(f"   📊 P-value: {result['p_value']:.4f}")
    print(f"   🎪 Decision: {result['decision']}")
    print(f"   🚨 Significant: {'Yes' if result['significant'] else 'No'}")

print(f"\n🎪 Hypothesis testing complete for all years! 🎪")
print(f"📊 Total tests performed: {len(all_results)}")
print(f"🎯 Significant results: {sum(1 for r in all_results if r['significant'])}")
print(f"🚨 Non-significant results: {sum(1 for r in all_results if not r['significant'])}")
```

```python
# 🎨 Step 3: Visualizing Our Hypothesis Test Results! 🎨
print("🎨 Step 3: Creating beautiful visualizations of our hypothesis test results!")
print("🌟 Let's see the patterns in our statistical investigation! 🌟")

# 🎨 Create comprehensive visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('🔍 Statistical Detective Investigation Results: Global Sugar Consumption 🔍', fontsize=20, fontweight='bold', y=0.98)

# 🎯 Extract data for plotting
years = [r['year'] for r in all_results]
sample_means = [r['sample_mean'] for r in all_results]
differences = [r['difference'] for r in all_results]
p_values = [r['p_value'] for r in all_results]
z_scores = [r['z_score'] for r in all_results]
significant = [r['significant'] for r in all_results]

# 🎨 Plot 1: Sample means over time
axes[0, 0].plot(years, sample_means, 'o-', color=detective_colors[0], linewidth=3, markersize=8, label='Sample Means')
axes[0, 0].axhline(mu_baseline, color='red', linestyle='--', linewidth=3, label=f'Baseline (1961): {mu_baseline:.1f}')
axes[0, 0].set_title('📊 Sample Means Over Time 📊', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Year', fontsize=14)
axes[0, 0].set_ylabel('Sugar Consumption (calories/day)', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# 🎨 Plot 2: Differences from baseline
colors = [detective_colors[1] if sig else detective_colors[2] for sig in significant]
axes[0, 1].bar(years, differences, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
axes[0, 1].axhline(0, color='red', linestyle='--', linewidth=2, alpha=0.7)
axes[0, 1].set_title('📈 Differences from Baseline (1961) 📈', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Year', fontsize=14)
axes[0, 1].set_ylabel('Difference (calories/day)', fontsize=14)
axes[0, 1].grid(True, alpha=0.3)

# 🎨 Plot 3: P-values over time
axes[1, 0].plot(years, p_values, 'o-', color=detective_colors[3], linewidth=3, markersize=8, label='P-values')
axes[1, 0].axhline(alpha, color='red', linestyle='--', linewidth=3, label=f'α = {alpha:.3f}')
axes[1, 0].set_title('📊 P-values Over Time 📊', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Year', fontsize=14)
axes[1, 0].set_ylabel('P-value', fontsize=14)
axes[1, 0].set_yscale('log')
axes[1, 0].legend(fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# 🎨 Plot 4: Z-scores over time
axes[1, 1].plot(years, z_scores, 'o-', color=detective_colors[4], linewidth=3, markersize=8, label='Z-scores')
axes[1, 1].axhline(0, color='red', linestyle='--', linewidth=2, alpha=0.7)
axes[1, 1].axhline(1.96, color='orange', linestyle=':', linewidth=2, alpha=0.7, label='±1.96 (α=0.05)')
axes[1, 1].axhline(-1.96, color='orange', linestyle=':', linewidth=2, alpha=0.7)
axes[1, 1].set_title('🧮 Z-scores Over Time 🧮', fontsize=16, fontweight='bold')
axes[1, 1].set_xlabel('Year', fontsize=14)
axes[1, 1].set_ylabel('Z-score', fontsize=14)
axes[1, 1].legend(fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 🎯 Summary statistics
print("\n🎯 Investigation Summary:")
print("🌟" + "="*60 + "🌟")
print(f"📊 Total years tested: {len(all_results)}")
print(f"🎯 Significant results: {sum(1 for r in all_results if r['significant'])}")
print(f"🚨 Non-significant results: {sum(1 for r in all_results if not r['significant'])}")
print(f"📈 Average difference: {np.mean(differences):.2f} calories/day")
print(f"📊 Average p-value: {np.mean(p_values):.4f}")
print(f"🧮 Average |Z-score|: {np.mean([abs(z) for z in z_scores]):.3f}")

# 🎯 Identify significant years
significant_years = [r['year'] for r in all_results if r['significant']]
if significant_years:
    print(f"\n🎯 Years with significant changes:")
    for year in significant_years:
        result = next(r for r in all_results if r['year'] == year)
        print(f"   📅 {year}: p-value = {result['p_value']:.4f}, difference = {result['difference']:.2f} cal/day")
else:
    print(f"\n🚨 No years showed significant changes at α = {alpha:.3f}")

print("🌟" + "="*60 + "🌟")
print("🎪 Our statistical investigation is complete! 🎪")
```

```python
# 🎪 Step 4: Sample Size Experiments! 🎪
print("🎪 Step 4: Investigating how sample size affects our conclusions!")
print("🔍 Let's see how sample size impacts statistical power! 🔍")

# 🎯 Find a year where we failed to reject H₀ (non-significant)
non_sig_results = [r for r in all_results if not r['significant']]
sig_results = [r for r in all_results if r['significant']]

print(f"\n📊 Sample Size Experiment Setup:")
print(f"   🚨 Non-significant results available: {len(non_sig_results)}")
print(f"   🎯 Significant results available: {len(sig_results)}")

# 🎪 Experiment 1: Decrease sample size for non-significant result
if non_sig_results:
    # 🎯 Pick the first non-significant result
    test_result = non_sig_results[0]
    test_year = test_result['year']
    
    print(f"\n🔍 EXPERIMENT 1: Decreasing Sample Size")
    print(f"🎯 Testing year {test_year} with smaller sample size")
    print(f"📊 Original sample size: {sample_size}")
    print(f"📊 New sample size: {sample_size//2}")
    
    # 📊 Get data for this year
    year_data = df[df.Year == test_year][category].dropna()
    
    if len(year_data) >= sample_size//2:
        # 🎲 Take a smaller random sample
        small_sample = random.sample(list(year_data), sample_size//2)
        
        # 🧮 Run hypothesis test with smaller sample
        small_result = hypothesis_test_detective(
            sample=small_sample,
            sample_size=sample_size//2,
            mu_baseline=mu_baseline,
            sigma_baseline=sigma_baseline,
            alpha=alpha,
            year=test_year
        )
        
        print(f"\n🎯 Sample Size Comparison for {test_year}:")
        print(f"   📊 Original (n={sample_size}): p-value = {test_result['p_value']:.4f}, decision = {test_result['decision']}")
        print(f"   📊 Smaller (n={sample_size//2}): p-value = {small_result['p_value']:.4f}, decision = {small_result['decision']}")
        print(f"   🎪 Effect: {'More significant' if small_result['p_value'] < test_result['p_value'] else 'Less significant'}")

# 🎪 Experiment 2: Increase sample size for significant result
if sig_results:
    # 🎯 Pick the first significant result
    test_result = sig_results[0]
    test_year = test_result['year']
    
    print(f"\n🔍 EXPERIMENT 2: Increasing Sample Size")
    print(f"🎯 Testing year {test_year} with larger sample size")
    print(f"📊 Original sample size: {sample_size}")
    print(f"📊 New sample size: {sample_size*2}")
    
    # 📊 Get data for this year
    year_data = df[df.Year == test_year][category].dropna()
    
    if len(year_data) >= sample_size*2:
        # 🎲 Take a larger random sample
        large_sample = random.sample(list(year_data), sample_size*2)
        
        # 🧮 Run hypothesis test with larger sample
        large_result = hypothesis_test_detective(
            sample=large_sample,
            sample_size=sample_size*2,
            mu_baseline=mu_baseline,
            sigma_baseline=sigma_baseline,
            alpha=alpha,
            year=test_year
        )
        
        print(f"\n🎯 Sample Size Comparison for {test_year}:")
        print(f"   📊 Original (n={sample_size}): p-value = {test_result['p_value']:.4f}, decision = {test_result['decision']}")
        print(f"   📊 Larger (n={sample_size*2}): p-value = {large_result['p_value']:.4f}, decision = {large_result['decision']}")
        print(f"   🎪 Effect: {'More significant' if large_result['p_value'] < test_result['p_value'] else 'Less significant'}")

print(f"\n🌟 Sample Size Experiments Complete!")
print(f"🎪 Key Insight: Larger sample sizes generally increase statistical power!")
print(f"🔍 This means we're more likely to detect real differences with larger samples!")
```
