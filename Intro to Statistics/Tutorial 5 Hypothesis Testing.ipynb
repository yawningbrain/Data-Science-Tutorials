## ðŸ” Understanding Hypothesis Testing: The Detective's Method ðŸ”

### ðŸŽ¯ What is Hypothesis Testing?
**Hypothesis testing** is a systematic way to make decisions about population parameters based on sample data. It's like being a detective who must decide whether evidence supports a claim or if it could have happened by chance.

### ðŸŽª The Two Competing Hypotheses:

#### ðŸ›¡ï¸ **Null Hypothesis (Hâ‚€)**: "Innocent Until Proven Guilty"
- **Definition**: A statement of **no difference** or **no effect**
- **Purpose**: The default assumption we test against
- **Examples**: 
  - "The new drug has no effect on blood pressure"
  - "There's no difference in sugar consumption between 1961 and 2013"
  - "The coin is fair (50% heads, 50% tails)"

#### ðŸŽ¯ **Alternative Hypothesis (Hâ‚)**: "The Claim We're Investigating"
- **Definition**: A claim that **contradicts** the null hypothesis
- **Purpose**: What we conclude if we reject Hâ‚€
- **Examples**:
  - "The new drug reduces blood pressure"
  - "Sugar consumption has changed since 1961"
  - "The coin is biased"

### ðŸ§® The Decision Process:
1. **ðŸ” Collect Evidence**: Gather sample data
2. **ðŸ§® Calculate Test Statistic**: Measure how extreme our sample is
3. **ðŸ“Š Find P-value**: Probability of seeing this result by chance
4. **ðŸŽ¯ Make Decision**: Compare p-value to significance level (Î±)

### ðŸŽª Decision Options:
- **"Reject Hâ‚€"**: Evidence strongly supports the alternative hypothesis
- **"Fail to Reject Hâ‚€"**: Evidence is insufficient to reject the null hypothesis
- **Note**: We never "accept" Hâ‚€ - we either reject it or fail to reject it!

### ðŸŽ¯ Significance Level (Î±): Our Threshold for "Reasonable Doubt"
- **Definition**: The probability of rejecting Hâ‚€ when it's actually true
- **Common Values**: Î± = 0.05 (5%), Î± = 0.01 (1%), Î± = 0.10 (10%)
- **Interpretation**: "We're willing to be wrong Î±% of the time"

### ðŸš¨ The Two Types of Errors:

#### ðŸš¨ **Type I Error (False Positive)**: "Crying Wolf"
- **Definition**: Rejecting a **true** null hypothesis
- **Probability**: Î± (our significance level)
- **Example**: Concluding a drug works when it doesn't
- **Risk**: Wasting resources on ineffective treatments

#### ðŸš¨ **Type II Error (False Negative)**: "Missing the Signal"
- **Definition**: Failing to reject a **false** null hypothesis
- **Probability**: Î² (depends on sample size and effect size)
- **Example**: Missing that a drug actually works
- **Risk**: Missing opportunities for effective treatments

### ðŸŽ¨ The Error Trade-off:
- **Decreasing Î±**: Reduces Type I errors but increases Type II errors
- **Increasing Sample Size**: Reduces both Type I and Type II errors
- **Larger Effect Size**: Reduces Type II errors

```python
# ðŸŒŸ Welcome to the Statistical Detective Agency! Let's set up our tools! ðŸŒŸ
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad 
from scipy.stats import norm, sem
import random
import warnings
warnings.filterwarnings('ignore')

# ðŸŽ¨ Let's make our plots beautiful and detective-worthy! âœ¨
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3

# ðŸŽ¯ Set up our detective color palette
detective_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F']

print("ðŸŒŸ Welcome to the Statistical Detective Agency! ðŸŒŸ")
print("âœ¨ All detective tools loaded successfully! âœ¨")
print("ðŸ” Ready to investigate statistical mysteries! ðŸ”")
print("ðŸŽª Let's solve this hypothesis testing case! ðŸŽª")
```

```python
# ðŸ§® Creating Our Detective's Hypothesis Testing Function! ðŸ§®
def standard_normal(x):
    """
    ðŸŽ¯ Standard normal probability density function
    
    Parameters:
    -----------
    x : float
        Value to evaluate
        
    Returns:
    --------
    float : PDF value at x
    """
    f_x = (1 / np.sqrt(2*np.pi)) * np.exp(-(x**2) / 2)
    return f_x

def hypothesis_test_detective(sample, sample_size, mu_baseline, sigma_baseline, alpha, year):
    """
    ðŸ” Statistical Detective's Hypothesis Testing Function
    
    Parameters:
    -----------
    sample : array-like
        Sample data from the year being tested
    sample_size : int
        Size of the sample
    mu_baseline : float
        Population mean from baseline year (1961)
    sigma_baseline : float
        Population standard deviation from baseline year
    alpha : float
        Significance level (e.g., 0.05 for 5%)
    year : int
        Year being tested
        
    Returns:
    --------
    dict : Results of the hypothesis test
    """
    
    # ðŸŽ¯ Calculate sample statistics
    x_bar = np.mean(sample)
    z_score = (x_bar - mu_baseline) / (sigma_baseline / np.sqrt(sample_size))
    
    # ðŸŽª Display hypotheses
    print(f"ðŸ”" + "="*60 + "ðŸ”")
    print(f"ðŸ” HYPOTHESIS TEST FOR YEAR {year} ðŸ”")
    print(f"ðŸ”" + "="*60 + "ðŸ”")
    print(f"ðŸ›¡ï¸ Hâ‚€ (Null Hypothesis): Global sugar consumption has NOT changed since 1961")
    print(f"   ðŸ“Š Baseline mean (1961): {mu_baseline:.2f} calories/day")
    print(f"ðŸŽ¯ Hâ‚ (Alternative Hypothesis): Global sugar consumption HAS changed since 1961")
    print(f"   ðŸ“Š Sample mean ({year}): {x_bar:.2f} calories/day")
    print(f"   ðŸ“ˆ Difference: {x_bar - mu_baseline:.2f} calories/day")
    print(f"   ðŸ§® Z-score: {z_score:.3f}")
    
    # ðŸ§® Calculate p-value (two-tailed test)
    if z_score >= 0:
        # Right tail
        p_val = 2 * (1 - norm.cdf(z_score))
    else:
        # Left tail
        p_val = 2 * norm.cdf(z_score)
    
    print(f"\nðŸ“Š Statistical Analysis:")
    print(f"   ðŸŽ¯ Sample size: {sample_size}")
    print(f"   ðŸ“ˆ P-value: {p_val:.4f} ({p_val*100:.2f}%)")
    print(f"   ðŸŽª Significance level (Î±): {alpha:.3f} ({alpha*100:.1f}%)")
    
    # ðŸŽ¯ Make decision
    if p_val < alpha:
        decision = "REJECT Hâ‚€"
        error_type = "Type I Error"
        error_description = "We conclude consumption has changed when it actually hasn't"
        print(f"\nðŸš¨ DECISION: {decision}")
        print(f"   âœ… P-value ({p_val:.4f}) < Î± ({alpha:.3f})")
        print(f"   ðŸŽ¯ Conclusion: Evidence suggests sugar consumption HAS changed since 1961")
    else:
        decision = "FAIL TO REJECT Hâ‚€"
        error_type = "Type II Error"
        error_description = "We conclude consumption hasn't changed when it actually has"
        print(f"\nðŸš¨ DECISION: {decision}")
        print(f"   âœ… P-value ({p_val:.4f}) â‰¥ Î± ({alpha:.3f})")
        print(f"   ðŸŽ¯ Conclusion: Insufficient evidence to claim sugar consumption has changed")
    
    print(f"\nðŸš¨ Potential Error Analysis:")
    print(f"   ðŸŽª Error Type: {error_type}")
    print(f"   ðŸ“Š Description: {error_description}")
    print(f"   ðŸŽ¯ Probability: {alpha:.1%} (Type I) or depends on effect size (Type II)")
    
    # ðŸŽª Return results
    results = {
        'year': year,
        'sample_mean': x_bar,
        'baseline_mean': mu_baseline,
        'difference': x_bar - mu_baseline,
        'z_score': z_score,
        'p_value': p_val,
        'alpha': alpha,
        'decision': decision,
        'error_type': error_type,
        'significant': p_val < alpha
    }
    
    return results

print("âœ… Detective's hypothesis testing function created successfully!")
print("ðŸ” Ready to investigate statistical mysteries! ðŸ”")
```

```python
# ðŸŒ Loading Our Global Nutrition Dataset! ðŸŒ
print("ðŸŒ Loading global dietary consumption data...")
print("ðŸŽª This dataset contains nutrition data from countries worldwide! ðŸŽª")

# ðŸ“Š Load the dataset
df = pd.read_csv('./Datasets/dietary-compositions-by-commodity-group.csv')

print("âœ… Dataset loaded successfully!")
print(f"ðŸ“Š Dataset shape: {df.shape}")
print(f"ðŸŒ Columns available: {list(df.columns)}")
print(f"ðŸ“… Years in dataset: {sorted(df['Year'].unique())}")
print(f"ðŸ›ï¸ Number of countries: {df['Entity'].nunique()}")

# ðŸŽ¨ Let's take a peek at our data
print("\nðŸŽ¨ First few rows of our dataset:")
display(df.head())

print("\nðŸŽ¨ Dataset info:")
print(df.info())

print("\nðŸŽ¨ Basic statistics:")
display(df.describe())

print("ðŸŒŸ Our global nutrition dataset is ready for investigation! ðŸŒŸ")
```

```python
# ðŸŽ¯ Step 1: Setting Up Our Investigation Parameters! ðŸŽ¯
print("ðŸŽ¯ Step 1: Setting up our statistical investigation!")
print("ðŸ” We'll investigate sugar consumption changes over time! ðŸ”")

# ðŸ­ Choose our food category
category = 'Sugar (FAO (2017)) (kilocalories per person per day)'
year_baseline = 1961

print(f"\nðŸ“Š Investigation Setup:")
print(f"   ðŸ­ Food Category: Sugar consumption")
print(f"   ðŸ“… Baseline Year: {year_baseline}")
print(f"   ðŸŽ¯ Population Parameter: Global average sugar consumption in {year_baseline}")

# ðŸ“Š Extract baseline data (1961)
df_baseline = df[df.Year == year_baseline]
baseline_data = df_baseline[category].dropna()

mu_baseline = np.mean(baseline_data)
sigma_baseline = np.std(baseline_data)

print(f"\nðŸ“Š Baseline Statistics ({year_baseline}):")
print(f"   ðŸŒ Number of countries: {len(baseline_data)}")
print(f"   ðŸ“ˆ Mean consumption: {mu_baseline:.2f} calories/day")
print(f"   ðŸ“Š Standard deviation: {sigma_baseline:.2f} calories/day")
print(f"   ðŸŽ¯ Min consumption: {np.min(baseline_data):.2f} calories/day")
print(f"   ðŸŽ¯ Max consumption: {np.max(baseline_data):.2f} calories/day")

# ðŸŽª Set up investigation parameters
years_to_test = np.arange(1962, 2013, 5)
alpha = 0.05  # 5% significance level
sample_size = 50

print(f"\nðŸŽª Investigation Parameters:")
print(f"   ðŸ“… Years to test: {list(years_to_test)}")
print(f"   ðŸŽ¯ Significance level (Î±): {alpha:.3f} ({alpha*100:.1f}%)")
print(f"   ðŸ“Š Sample size: {sample_size} countries per test")

# ðŸŽ¨ Visualize baseline distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('ðŸ­ Baseline Sugar Consumption Distribution (1961) ðŸ­', fontsize=16, fontweight='bold')

# ðŸŽ¨ Plot 1: Histogram
axes[0].hist(baseline_data, bins=30, color=detective_colors[0], alpha=0.8, edgecolor='black')
axes[0].axvline(mu_baseline, color='red', linestyle='--', linewidth=3, label=f'Mean: {mu_baseline:.1f}')
axes[0].set_title('ðŸ“Š Distribution of Sugar Consumption', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sugar Consumption (calories/day)', fontsize=12)
axes[0].set_ylabel('Number of Countries', fontsize=12)
axes[0].legend(fontsize=12)
axes[0].grid(True, alpha=0.3)

# ðŸŽ¨ Plot 2: Box plot
bp = axes[1].boxplot(baseline_data, vert=True, patch_artist=True,
                    boxprops=dict(facecolor=detective_colors[1], alpha=0.8, linewidth=2),
                    medianprops=dict(color='red', linewidth=3),
                    whiskerprops=dict(linewidth=2),
                    capprops=dict(linewidth=2))
axes[1].set_title('ðŸ“¦ Box Plot of Sugar Consumption', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Sugar Consumption (calories/day)', fontsize=12)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("ðŸŒŸ Baseline analysis complete! Ready to test for changes over time! ðŸŒŸ")
```

```python
# ðŸŽª Step 2: Running Our Hypothesis Tests! ðŸŽª
print("ðŸŽª Step 2: Running hypothesis tests for each year!")
print("ðŸ” Let's see which years show significant changes in sugar consumption! ðŸ”")

# ðŸŽ¯ Store results for analysis
all_results = []

# ðŸŽª Test each year
for year in years_to_test:
    print(f"\n{'='*80}")
    print(f"ðŸ” TESTING YEAR {year} ðŸ”")
    print(f"{'='*80}")
    
    # ðŸ“Š Get data for this year
    year_data = df[df.Year == year][category].dropna()
    
    if len(year_data) < sample_size:
        print(f"âš ï¸ Warning: Only {len(year_data)} countries available for {year}")
        print(f"   Skipping this year due to insufficient data.")
        continue
    
    # ðŸŽ² Take a random sample
    sample = random.sample(list(year_data), sample_size)
    
    # ðŸ§® Run hypothesis test
    result = hypothesis_test_detective(
        sample=sample,
        sample_size=sample_size,
        mu_baseline=mu_baseline,
        sigma_baseline=sigma_baseline,
        alpha=alpha,
        year=year
    )
    
    # ðŸŽ¯ Store results
    all_results.append(result)
    
    print(f"\nðŸŽ¯ Year {year} Summary:")
    print(f"   ðŸ“Š Sample mean: {result['sample_mean']:.2f} calories/day")
    print(f"   ðŸ“ˆ Difference from baseline: {result['difference']:.2f} calories/day")
    print(f"   ðŸ§® Z-score: {result['z_score']:.3f}")
    print(f"   ðŸ“Š P-value: {result['p_value']:.4f}")
    print(f"   ðŸŽª Decision: {result['decision']}")
    print(f"   ðŸš¨ Significant: {'Yes' if result['significant'] else 'No'}")

print(f"\nðŸŽª Hypothesis testing complete for all years! ðŸŽª")
print(f"ðŸ“Š Total tests performed: {len(all_results)}")
print(f"ðŸŽ¯ Significant results: {sum(1 for r in all_results if r['significant'])}")
print(f"ðŸš¨ Non-significant results: {sum(1 for r in all_results if not r['significant'])}")
```

```python
# ðŸŽ¨ Step 3: Visualizing Our Hypothesis Test Results! ðŸŽ¨
print("ðŸŽ¨ Step 3: Creating beautiful visualizations of our hypothesis test results!")
print("ðŸŒŸ Let's see the patterns in our statistical investigation! ðŸŒŸ")

# ðŸŽ¨ Create comprehensive visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('ðŸ” Statistical Detective Investigation Results: Global Sugar Consumption ðŸ”', fontsize=20, fontweight='bold', y=0.98)

# ðŸŽ¯ Extract data for plotting
years = [r['year'] for r in all_results]
sample_means = [r['sample_mean'] for r in all_results]
differences = [r['difference'] for r in all_results]
p_values = [r['p_value'] for r in all_results]
z_scores = [r['z_score'] for r in all_results]
significant = [r['significant'] for r in all_results]

# ðŸŽ¨ Plot 1: Sample means over time
axes[0, 0].plot(years, sample_means, 'o-', color=detective_colors[0], linewidth=3, markersize=8, label='Sample Means')
axes[0, 0].axhline(mu_baseline, color='red', linestyle='--', linewidth=3, label=f'Baseline (1961): {mu_baseline:.1f}')
axes[0, 0].set_title('ðŸ“Š Sample Means Over Time ðŸ“Š', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Year', fontsize=14)
axes[0, 0].set_ylabel('Sugar Consumption (calories/day)', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# ðŸŽ¨ Plot 2: Differences from baseline
colors = [detective_colors[1] if sig else detective_colors[2] for sig in significant]
axes[0, 1].bar(years, differences, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
axes[0, 1].axhline(0, color='red', linestyle='--', linewidth=2, alpha=0.7)
axes[0, 1].set_title('ðŸ“ˆ Differences from Baseline (1961) ðŸ“ˆ', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Year', fontsize=14)
axes[0, 1].set_ylabel('Difference (calories/day)', fontsize=14)
axes[0, 1].grid(True, alpha=0.3)

# ðŸŽ¨ Plot 3: P-values over time
axes[1, 0].plot(years, p_values, 'o-', color=detective_colors[3], linewidth=3, markersize=8, label='P-values')
axes[1, 0].axhline(alpha, color='red', linestyle='--', linewidth=3, label=f'Î± = {alpha:.3f}')
axes[1, 0].set_title('ðŸ“Š P-values Over Time ðŸ“Š', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Year', fontsize=14)
axes[1, 0].set_ylabel('P-value', fontsize=14)
axes[1, 0].set_yscale('log')
axes[1, 0].legend(fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# ðŸŽ¨ Plot 4: Z-scores over time
axes[1, 1].plot(years, z_scores, 'o-', color=detective_colors[4], linewidth=3, markersize=8, label='Z-scores')
axes[1, 1].axhline(0, color='red', linestyle='--', linewidth=2, alpha=0.7)
axes[1, 1].axhline(1.96, color='orange', linestyle=':', linewidth=2, alpha=0.7, label='Â±1.96 (Î±=0.05)')
axes[1, 1].axhline(-1.96, color='orange', linestyle=':', linewidth=2, alpha=0.7)
axes[1, 1].set_title('ðŸ§® Z-scores Over Time ðŸ§®', fontsize=16, fontweight='bold')
axes[1, 1].set_xlabel('Year', fontsize=14)
axes[1, 1].set_ylabel('Z-score', fontsize=14)
axes[1, 1].legend(fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ðŸŽ¯ Summary statistics
print("\nðŸŽ¯ Investigation Summary:")
print("ðŸŒŸ" + "="*60 + "ðŸŒŸ")
print(f"ðŸ“Š Total years tested: {len(all_results)}")
print(f"ðŸŽ¯ Significant results: {sum(1 for r in all_results if r['significant'])}")
print(f"ðŸš¨ Non-significant results: {sum(1 for r in all_results if not r['significant'])}")
print(f"ðŸ“ˆ Average difference: {np.mean(differences):.2f} calories/day")
print(f"ðŸ“Š Average p-value: {np.mean(p_values):.4f}")
print(f"ðŸ§® Average |Z-score|: {np.mean([abs(z) for z in z_scores]):.3f}")

# ðŸŽ¯ Identify significant years
significant_years = [r['year'] for r in all_results if r['significant']]
if significant_years:
    print(f"\nðŸŽ¯ Years with significant changes:")
    for year in significant_years:
        result = next(r for r in all_results if r['year'] == year)
        print(f"   ðŸ“… {year}: p-value = {result['p_value']:.4f}, difference = {result['difference']:.2f} cal/day")
else:
    print(f"\nðŸš¨ No years showed significant changes at Î± = {alpha:.3f}")

print("ðŸŒŸ" + "="*60 + "ðŸŒŸ")
print("ðŸŽª Our statistical investigation is complete! ðŸŽª")
```
