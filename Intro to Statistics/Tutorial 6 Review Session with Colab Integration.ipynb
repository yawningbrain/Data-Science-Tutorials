```python
# üé≤ Section 3: Binomial Distribution Mastery! üé≤
print("üé≤ Section 3: Binomial Distribution Mastery!")
print("üé™ Let's demonstrate probability distributions and simulations! üé™")

# üéØ Binomial distribution simulation parameters
N = 20000  # Number of trials
n = 10     # Number of flips per trial

# üé™ Different probability scenarios
p0 = 0.5   # Fair coin
p1 = 0.25  # Biased coin (heads less likely)
p2 = 0.75  # Biased coin (heads more likely)

print(f"\nüéØ Simulation Parameters:")
print(f"   üé≤ Number of trials: {N:,}")
print(f"   ü™ô Flips per trial: {n}")
print(f"   üé™ Probability scenarios:")
print(f"      - Fair coin (p=0.5): {p0}")
print(f"      - Biased coin (p=0.25): {p1}")
print(f"      - Biased coin (p=0.75): {p2}")

# üéØ Initialize result lists
result_p0 = []  # Fair coin results
result_p1 = []  # Biased coin (low p) results
result_p2 = []  # Biased coin (high p) results

print(f"\nüé™ Running binomial simulations...")
print(f"   üé≤ Simulating {N:,} trials for each probability...")

# üé™ Run the binomial simulations
for i in range(N):
    # üé≤ Fair coin simulation
    trial_p0 = (np.random.rand(n) <= p0)  # Boolean array: True = Heads, False = Tails
    result_p0.append(np.sum(trial_p0))    # Count heads in this trial
    
    # üé≤ Biased coin (low p) simulation
    trial_p1 = (np.random.rand(n) <= p1)  # Boolean array: True = Heads, False = Tails
    result_p1.append(np.sum(trial_p1))    # Count heads in this trial
    
    # üé≤ Biased coin (high p) simulation
    trial_p2 = (np.random.rand(n) <= p2)  # Boolean array: True = Heads, False = Tails
    result_p2.append(np.sum(trial_p2))    # Count heads in this trial
    
    # üé™ Progress indicator
    if (i + 1) % 5000 == 0:
        print(f"      ‚úÖ Completed {i + 1:,} trials...")

print(f"‚úÖ Binomial simulations complete!")
print(f"üìä Results summary:")
print(f"   üé≤ Fair coin (p=0.5): Mean = {np.mean(result_p0):.2f}, Std = {np.std(result_p0):.2f}")
print(f"   üé≤ Biased coin (p=0.25): Mean = {np.mean(result_p1):.2f}, Std = {np.std(result_p1):.2f}")
print(f"   üé≤ Biased coin (p=0.75): Mean = {np.mean(result_p2):.2f}, Std = {np.std(result_p2):.2f}")

# üéØ Theoretical expectations
print(f"\nüéØ Theoretical Expectations:")
print(f"   üé≤ Fair coin (p=0.5): Expected mean = {n * p0:.1f}, Expected std = {np.sqrt(n * p0 * (1 - p0)):.2f}")
print(f"   üé≤ Biased coin (p=0.25): Expected mean = {n * p1:.1f}, Expected std = {np.sqrt(n * p1 * (1 - p1)):.2f}")
print(f"   üé≤ Biased coin (p=0.75): Expected mean = {n * p2:.1f}, Expected std = {np.sqrt(n * p2 * (1 - p2)):.2f}")

print("üåü Binomial distribution mastery demonstrated! üåü")
```

```python
# üé® Mastery Challenge 3: Binomial Distribution Visualization! üé®
print("üé® Mastery Challenge 3: Binomial Distribution Visualization!")
print("üåü Let's create stunning visualizations of our binomial distributions! üåü")

# üé® Create comprehensive binomial distribution visualization
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('üé≤ Binomial Distribution Mastery: Experimental vs Theoretical! üé≤', fontsize=20, fontweight='bold', y=0.98)

# üéØ Define our probability scenarios
probabilities = [p0, p1, p2]
results = [result_p0, result_p1, result_p2]
colors = [mastery_colors[0], mastery_colors[1], mastery_colors[2]]
labels = ['Fair Coin (p=0.5)', 'Biased Coin (p=0.25)', 'Biased Coin (p=0.75)']

# üé™ Create histograms and theoretical comparisons
for i, (p, result, color, label) in enumerate(zip(probabilities, results, colors, labels)):
    
    # üé® Plot 1: Experimental histogram
    axes[0, i].hist(result, bins=50, color=color, alpha=0.8, density=True, edgecolor='black', label=f'Experimental (n={N:,})')
    
    # üéØ Add theoretical binomial distribution
    x_range = np.arange(0, n+1)
    theoretical_probs = [binom.pmf(k, n, p) for k in x_range]
    axes[0, i].plot(x_range, theoretical_probs, 'ro-', linewidth=3, markersize=8, label='Theoretical')
    
    axes[0, i].set_title(f'üé≤ {label} üé≤', fontsize=16, fontweight='bold')
    axes[0, i].set_xlabel('Number of Heads', fontsize=14)
    axes[0, i].set_ylabel('Probability', fontsize=14)
    axes[0, i].legend(fontsize=12)
    axes[0, i].grid(True, alpha=0.3)
    
    # üé® Plot 2: Comparison of means
    experimental_mean = np.mean(result)
    theoretical_mean = n * p
    experimental_std = np.std(result)
    theoretical_std = np.sqrt(n * p * (1 - p))
    
    means = [experimental_mean, theoretical_mean]
    stds = [experimental_std, theoretical_std]
    categories = ['Experimental', 'Theoretical']
    
    bars = axes[1, i].bar(categories, means, color=[color, 'red'], alpha=0.8, edgecolor='black')
    axes[1, i].errorbar(categories, means, yerr=stds, fmt='none', color='black', capsize=10, capthick=2)
    
    # üéØ Add value labels on bars
    for j, (mean, std) in enumerate(zip(means, stds)):
        axes[1, i].text(j, mean + std + 0.1, f'Œº={mean:.2f}\nœÉ={std:.2f}', 
                       ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    axes[1, i].set_title(f'üìä Mean & Std Comparison: {label} üìä', fontsize=16, fontweight='bold')
    axes[1, i].set_ylabel('Value', fontsize=14)
    axes[1, i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# üéØ Statistical analysis
print("\nüéØ Binomial Distribution Analysis:")
print("üåü" + "="*60 + "üåü")

for i, (p, result, label) in enumerate(zip(probabilities, results, labels)):
    experimental_mean = np.mean(result)
    theoretical_mean = n * p
    experimental_std = np.std(result)
    theoretical_std = np.sqrt(n * p * (1 - p))
    
    print(f"\nüìä {label}:")
    print(f"   üé≤ Experimental: Mean = {experimental_mean:.3f}, Std = {experimental_std:.3f}")
    print(f"   üéØ Theoretical: Mean = {theoretical_mean:.3f}, Std = {theoretical_std:.3f}")
    print(f"   ‚úÖ Mean difference: {abs(experimental_mean - theoretical_mean):.4f}")
    print(f"   ‚úÖ Std difference: {abs(experimental_std - theoretical_std):.4f}")

print("\nüåü Key Insights:")
print("   üé™ Experimental results match theoretical predictions!")
print("   üìä Large sample sizes (N=20,000) provide excellent approximations")
print("   üéØ Binomial distribution shape depends on probability p")
print("   üßÆ Mean = n√óp, Variance = n√óp√ó(1-p)")

print("üåü" + "="*60 + "üåü")
print("üé™ Binomial distribution visualization mastery demonstrated! üé™")
```

```python
# üåü Section 4: Gaussian Distribution Mastery! üåü
print("üåü Section 4: Gaussian Distribution Mastery!")
print("üé™ Let's demonstrate the power of normal distributions! üé™")

# üéØ Gaussian distribution parameters
x = 20  # Maximum mean value
y = 10  # Maximum standard deviation

# üé≤ Generate random parameters
mu = np.round(x * np.random.rand(), 2)      # Random mean between 0 and 20
sigma = np.round(y * np.random.rand(), 2)   # Random std between 0 and 10

# üìä Generate Gaussian distribution
samples = 10000
gaussian_dist = sigma * np.random.randn(samples) + mu

print(f"\nüéØ Gaussian Distribution Parameters:")
print(f"   üåü Mean (Œº): {mu}")
print(f"   üìä Standard Deviation (œÉ): {sigma}")
print(f"   üìà Number of samples: {samples:,}")

# üé® Create comprehensive Gaussian distribution visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('üåü Gaussian Distribution Mastery: The Foundation of Statistics! üåü', fontsize=20, fontweight='bold', y=0.98)

# üé® Plot 1: Main histogram
axes[0, 0].hist(gaussian_dist, bins=50, color=mastery_colors[0], alpha=0.8, density=True, edgecolor='black', label='Experimental')
axes[0, 0].axvline(mu, color='red', linestyle='--', linewidth=3, label=f'Mean: {mu}')
axes[0, 0].set_title('üìä Gaussian Distribution Histogram üìä', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Value', fontsize=14)
axes[0, 0].set_ylabel('Density', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# üé® Plot 2: Theoretical vs Experimental
x_range = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
theoretical_pdf = norm.pdf(x_range, mu, sigma)
axes[0, 1].hist(gaussian_dist, bins=50, color=mastery_colors[1], alpha=0.6, density=True, edgecolor='black', label='Experimental')
axes[0, 1].plot(x_range, theoretical_pdf, 'r-', linewidth=3, label='Theoretical PDF')
axes[0, 1].set_title('üéØ Experimental vs Theoretical PDF üéØ', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Value', fontsize=14)
axes[0, 1].set_ylabel('Density', fontsize=14)
axes[0, 1].legend(fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# üé® Plot 3: Empirical Rule demonstration
axes[1, 0].hist(gaussian_dist, bins=50, color=mastery_colors[2], alpha=0.8, density=True, edgecolor='black')
axes[1, 0].axvline(mu, color='red', linestyle='-', linewidth=3, label=f'Mean: {mu}')
axes[1, 0].axvline(mu + sigma, color='orange', linestyle='--', linewidth=2, label=f'Œº + œÉ: {mu + sigma:.2f}')
axes[1, 0].axvline(mu - sigma, color='orange', linestyle='--', linewidth=2, label=f'Œº - œÉ: {mu - sigma:.2f}')
axes[1, 0].axvline(mu + 2*sigma, color='green', linestyle=':', linewidth=2, label=f'Œº + 2œÉ: {mu + 2*sigma:.2f}')
axes[1, 0].axvline(mu - 2*sigma, color='green', linestyle=':', linewidth=2, label=f'Œº - 2œÉ: {mu - 2*sigma:.2f}')
axes[1, 0].set_title('üìè Empirical Rule Demonstration üìè', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Value', fontsize=14)
axes[1, 0].set_ylabel('Density', fontsize=14)
axes[1, 0].legend(fontsize=10)
axes[1, 0].grid(True, alpha=0.3)

# üé® Plot 4: Box plot and statistics
bp = axes[1, 1].boxplot(gaussian_dist, vert=True, patch_artist=True,
                       boxprops=dict(facecolor=mastery_colors[3], alpha=0.8, linewidth=2),
                       medianprops=dict(color='red', linewidth=3),
                       whiskerprops=dict(linewidth=2),
                       capprops=dict(linewidth=2))
axes[1, 1].set_title('üì¶ Box Plot Analysis üì¶', fontsize=16, fontweight='bold')
axes[1, 1].set_ylabel('Value', fontsize=14)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# üéØ Statistical analysis
print("\nüéØ Gaussian Distribution Analysis:")
print("üåü" + "="*60 + "üåü")

# üìä Calculate statistics
experimental_mean = np.mean(gaussian_dist)
experimental_std = np.std(gaussian_dist)
experimental_median = np.median(gaussian_dist)

print(f"üìä Distribution Statistics:")
print(f"   üåü Experimental mean: {experimental_mean:.3f} (theoretical: {mu:.3f})")
print(f"   üìä Experimental std: {experimental_std:.3f} (theoretical: {sigma:.3f})")
print(f"   üìà Experimental median: {experimental_median:.3f}")
print(f"   ‚úÖ Mean difference: {abs(experimental_mean - mu):.4f}")
print(f"   ‚úÖ Std difference: {abs(experimental_std - sigma):.4f}")

# üéØ Empirical Rule verification
within_1sigma = np.sum((gaussian_dist >= mu - sigma) & (gaussian_dist <= mu + sigma)) / len(gaussian_dist)
within_2sigma = np.sum((gaussian_dist >= mu - 2*sigma) & (gaussian_dist <= mu + 2*sigma)) / len(gaussian_dist)
within_3sigma = np.sum((gaussian_dist >= mu - 3*sigma) & (gaussian_dist <= mu + 3*sigma)) / len(gaussian_dist)

print(f"\nüìè Empirical Rule Verification:")
print(f"   üéØ Within 1œÉ: {within_1sigma:.1%} (expected: 68.3%)")
print(f"   üéØ Within 2œÉ: {within_2sigma:.1%} (expected: 95.4%)")
print(f"   üéØ Within 3œÉ: {within_3sigma:.1%} (expected: 99.7%)")

print("\nüåü Key Insights:")
print("   üé™ Gaussian distributions are symmetric around the mean")
print("   üìä Empirical Rule helps identify outliers")
print("   üéØ Normal distributions are fundamental to statistical inference")
print("   üßÆ Central Limit Theorem relies on normal distributions")

print("üåü" + "="*60 + "üåü")
print("üé™ Gaussian distribution mastery demonstrated! üé™")
```

```python
# üßÆ Section 5: Advanced Gaussian Analysis - Integration Mastery! üßÆ
print("üßÆ Section 5: Advanced Gaussian Analysis - Integration Mastery!")
print("üé™ Let's demonstrate the mathematical beauty of Gaussian functions! üé™")

# üéØ Define Gaussian function
def gaussian(x, mu, sigma):
    """
    üåü Gaussian probability density function
    
    Parameters:
    -----------
    x : float or array
        Input values
    mu : float
        Mean of the distribution
    sigma : float
        Standard deviation of the distribution
        
    Returns:
    --------
    float or array : PDF values
    """
    f_x = (1 / np.sqrt(2*np.pi*sigma*sigma)) * np.exp(-(x - mu)*(x - mu) / (2*sigma*sigma))
    return f_x

# üéØ Standard normal distribution parameters
mu = 0.0  # Mean
sigma = 1.0  # Standard deviation

print(f"\nüéØ Gaussian Integration Analysis:")
print(f"   üåü Mean (Œº): {mu}")
print(f"   üìä Standard Deviation (œÉ): {sigma}")
print(f"   üé™ Distribution: Standard Normal")

# üé® Create comprehensive integration visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('üßÆ Gaussian Integration Mastery: The Mathematical Foundation! üßÆ', fontsize=20, fontweight='bold', y=0.98)

# üé® Plot 1: Full Gaussian curve
x_range = np.linspace(-4, 4, 1000)
y_values = gaussian(x_range, mu, sigma)
axes[0, 0].plot(x_range, y_values, color=mastery_colors[0], linewidth=3, label='Standard Normal PDF')
axes[0, 0].set_title('üåü Standard Normal Distribution üåü', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Z-score', fontsize=14)
axes[0, 0].set_ylabel('Probability Density', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# üé® Plot 2: Empirical Rule intervals
for n in range(1, 4):
    # üéØ Calculate integration
    I = quad(gaussian, -n*sigma, n*sigma, args=(mu, sigma))[0]
    
    # üé® Plot shaded areas
    x_shade = np.linspace(-n*sigma, n*sigma, 1000)
    y_shade = gaussian(x_shade, mu, sigma)
    axes[0, 1].fill_between(x_shade, y_shade, alpha=0.3, color=mastery_colors[n], 
                           label=f'{n}œÉ interval: {I*100:.1f}%')
    
    print(f"   üìä The {n}œÉ interval contains {np.round(I*100, 1)}% of the data.")

axes[0, 1].plot(x_range, y_values, color='black', linewidth=2, alpha=0.7)
axes[0, 1].set_title('üìè Empirical Rule: Integration Areas üìè', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Z-score', fontsize=14)
axes[0, 1].set_ylabel('Probability Density', fontsize=14)
axes[0, 1].legend(fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# üé® Plot 3: Cumulative Distribution Function
cdf_values = []
x_cdf = np.linspace(-4, 4, 100)
for x in x_cdf:
    cdf_val = quad(gaussian, -np.inf, x, args=(mu, sigma))[0]
    cdf_values.append(cdf_val)

axes[1, 0].plot(x_cdf, cdf_values, color=mastery_colors[2], linewidth=3, label='CDF (Integration)')
axes[1, 0].plot(x_cdf, norm.cdf(x_cdf), 'r--', linewidth=2, label='Scipy CDF')
axes[1, 0].set_title('üìà Cumulative Distribution Function üìà', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Z-score', fontsize=14)
axes[1, 0].set_ylabel('Cumulative Probability', fontsize=14)
axes[1, 0].legend(fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# üé® Plot 4: Integration comparison
intervals = ['1œÉ', '2œÉ', '3œÉ']
integrated_areas = []
scipy_areas = []

for n in range(1, 4):
    # üéØ Our integration
    I = quad(gaussian, -n*sigma, n*sigma, args=(mu, sigma))[0]
    integrated_areas.append(I * 100)
    
    # üéØ Scipy verification
    scipy_area = (norm.cdf(n*sigma) - norm.cdf(-n*sigma)) * 100
    scipy_areas.append(scipy_area)

x_pos = np.arange(len(intervals))
width = 0.35

bars1 = axes[1, 1].bar(x_pos - width/2, integrated_areas, width, label='Our Integration', 
                      color=mastery_colors[3], alpha=0.8, edgecolor='black')
bars2 = axes[1, 1].bar(x_pos + width/2, scipy_areas, width, label='Scipy CDF', 
                      color=mastery_colors[4], alpha=0.8, edgecolor='black')

axes[1, 1].set_title('üéØ Integration Verification üéØ', fontsize=16, fontweight='bold')
axes[1, 1].set_xlabel('Interval', fontsize=14)
axes[1, 1].set_ylabel('Percentage of Data (%)', fontsize=14)
axes[1, 1].set_xticks(x_pos)
axes[1, 1].set_xticklabels(intervals)
axes[1, 1].legend(fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

# üéØ Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width()/2., height + 0.5,
                   f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

for bar in bars2:
    height = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width()/2., height + 0.5,
                   f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# üéØ Advanced integration analysis
print(f"\nüéØ Advanced Integration Analysis:")
print("üåü" + "="*60 + "üåü")

# üéØ Test specific z-values
test_z_values = [0, 1, 1.96, 2.58]
print(f"\nüìä CDF Values for Specific Z-scores:")
for z in test_z_values:
    cdf_val = quad(gaussian, -np.inf, z, args=(mu, sigma))[0]
    scipy_val = norm.cdf(z)
    print(f"   üéØ Z = {z:4.2f}: Our integration = {cdf_val:.4f}, Scipy = {scipy_val:.4f}, Difference = {abs(cdf_val - scipy_val):.6f}")

# üéØ Percentile calculations
percentiles = [0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95]
print(f"\nüìà Percentile Calculations:")
for p in percentiles:
    # üéØ Find z-value for given percentile
    z_val = norm.ppf(p)
    print(f"   üé™ {p*100:4.0f}th percentile: Z = {z_val:6.3f}")

print("\nüåü Key Insights:")
print("   üßÆ Integration is fundamental to probability calculations")
print("   üìä Empirical Rule: 68-95-99.7% within 1-2-3 standard deviations")
print("   üéØ CDF provides cumulative probabilities")
print("   üåü Standard normal distribution is the foundation of statistical inference")

print("üåü" + "="*60 + "üåü")
print("üé™ Gaussian integration mastery demonstrated! üé™")
```

```python
# üéØ Section 6: CDF and ICDF Mastery - The Complete Picture! üéØ
print("üéØ Section 6: CDF and ICDF Mastery - The Complete Picture!")
print("üé™ Let's demonstrate the full power of cumulative functions! üé™")

# üéØ CDF Analysis
z = 1.25
print(f"\nüéØ Cumulative Distribution Function (CDF) Analysis:")
print(f"   üé™ Z-score: {z}")

# üßÆ Calculate CDF using integration
I = quad(gaussian, -np.inf, z, args=(mu, sigma))[0]
I_check = norm.cdf(z)

print(f"   üßÆ Integration method: F(Z < {z}) = {I:.4f}")
print(f"   üìä Scipy CDF function: F(Z < {z}) = {I_check:.4f}")
print(f"   ‚úÖ Difference: {abs(I - I_check):.6f}")

# üéØ ICDF Analysis
mean_gauss = 25
sigma_gauss = 4
p = 0.1

print(f"\nüéØ Inverse Cumulative Distribution Function (ICDF) Analysis:")
print(f"   üé™ Probability: {p*100}%")
print(f"   üåü Mean: {mean_gauss}")
print(f"   üìä Standard deviation: {sigma_gauss}")

# üßÆ Calculate ICDF
Q = norm.ppf(p)  # Standard normal quantile
Q_gauss = sigma_gauss * Q + mean_gauss  # Transform to our distribution

print(f"   üßÆ ICDF result: Q(p={p*100}%) = {Q_gauss:.2f}")

# üéØ Verification
check_p = quad(gaussian, -np.inf, Q_gauss, args=(mean_gauss, sigma_gauss))[0]

print(f"   ‚úÖ Verification: p = {check_p*100:.2f}% (should be {p*100:.1f}%)")
print(f"   ‚úÖ Difference: {abs(check_p - p)*100:.4f}%")

# üé® Create comprehensive CDF and ICDF visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('üéØ CDF and ICDF Mastery: The Complete Statistical Toolkit! üéØ', fontsize=20, fontweight='bold', y=0.98)

# üé® Plot 1: CDF demonstration
x_cdf = np.linspace(-4, 4, 1000)
y_cdf = norm.cdf(x_cdf)
axes[0, 0].plot(x_cdf, y_cdf, color=mastery_colors[0], linewidth=3, label='Standard Normal CDF')
axes[0, 0].axvline(z, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'Z = {z}')
axes[0, 0].axhline(I_check, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'F(Z < {z}) = {I_check:.3f}')
axes[0, 0].fill_between(x_cdf[x_cdf <= z], y_cdf[x_cdf <= z], alpha=0.3, color='red')
axes[0, 0].set_title('üìà Cumulative Distribution Function (CDF) üìà', fontsize=16, fontweight='bold')
axes[0, 0].set_xlabel('Z-score', fontsize=14)
axes[0, 0].set_ylabel('Cumulative Probability', fontsize=14)
axes[0, 0].legend(fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# üé® Plot 2: ICDF demonstration
p_range = np.linspace(0.01, 0.99, 1000)
q_range = norm.ppf(p_range)
axes[0, 1].plot(p_range, q_range, color=mastery_colors[1], linewidth=3, label='Standard Normal ICDF')
axes[0, 1].axhline(Q, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'Q(p={p*100}%) = {Q:.3f}')
axes[0, 1].axvline(p, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'p = {p:.3f}')
axes[0, 1].set_title('üìä Inverse Cumulative Distribution Function (ICDF) üìä', fontsize=16, fontweight='bold')
axes[0, 1].set_xlabel('Probability', fontsize=14)
axes[0, 1].set_ylabel('Z-score', fontsize=14)
axes[0, 1].legend(fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# üé® Plot 3: Transformed Gaussian
x_transformed = np.linspace(mean_gauss - 4*sigma_gauss, mean_gauss + 4*sigma_gauss, 1000)
y_transformed = gaussian(x_transformed, mean_gauss, sigma_gauss)
axes[1, 0].plot(x_transformed, y_transformed, color=mastery_colors[2], linewidth=3, label=f'Gaussian(Œº={mean_gauss}, œÉ={sigma_gauss})')
axes[1, 0].axvline(Q_gauss, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'Q(p={p*100}%) = {Q_gauss:.2f}')
axes[1, 0].fill_between(x_transformed[x_transformed <= Q_gauss], y_transformed[x_transformed <= Q_gauss], alpha=0.3, color='red')
axes[1, 0].set_title('üåü Transformed Gaussian Distribution üåü', fontsize=16, fontweight='bold')
axes[1, 0].set_xlabel('Value', fontsize=14)
axes[1, 0].set_ylabel('Probability Density', fontsize=14)
axes[1, 0].legend(fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# üé® Plot 4: CDF and ICDF relationship
axes[1, 1].plot(x_cdf, y_cdf, color=mastery_colors[3], linewidth=3, label='CDF')
axes[1, 1].plot(y_cdf, x_cdf, color=mastery_colors[4], linewidth=3, label='ICDF')
axes[1, 1].plot([0, 1], [0, 1], 'k--', alpha=0.5, label='y = x')
axes[1, 1].set_title('üé™ CDF and ICDF Relationship üé™', fontsize=16, fontweight='bold')
axes[1, 1].set_xlabel('Probability', fontsize=14)
axes[1, 1].set_ylabel('Z-score', fontsize=14)
axes[1, 1].legend(fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# üéØ Advanced CDF and ICDF analysis
print(f"\nüéØ Advanced CDF and ICDF Analysis:")
print("üåü" + "="*60 + "üåü")

# üéØ Test multiple percentiles
percentiles = [0.01, 0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99]
print(f"\nüìä Percentile Analysis:")
print(f"   {'Percentile':<12} {'Z-score':<10} {'Transformed':<12} {'Verification':<12}")
print(f"   {'-'*12} {'-'*10} {'-'*12} {'-'*12}")

for p in percentiles:
    z_val = norm.ppf(p)
    transformed_val = sigma_gauss * z_val + mean_gauss
    verification = quad(gaussian, -np.inf, transformed_val, args=(mean_gauss, sigma_gauss))[0]
    print(f"   {p*100:8.1f}%    {z_val:8.3f}    {transformed_val:8.2f}    {verification*100:8.2f}%")

# üéØ Confidence intervals
confidence_levels = [0.68, 0.95, 0.99]
print(f"\nüéØ Confidence Intervals for Transformed Gaussian:")
for conf in confidence_levels:
    alpha = 1 - conf
    z_lower = norm.ppf(alpha/2)
    z_upper = norm.ppf(1 - alpha/2)
    
    lower_bound = sigma_gauss * z_lower + mean_gauss
    upper_bound = sigma_gauss * z_upper + mean_gauss
    
    print(f"   üé™ {conf*100:.0f}% CI: [{lower_bound:.2f}, {upper_bound:.2f}]")

print("\nüåü Key Insights:")
print("   üìà CDF: P(X ‚â§ x) - cumulative probability")
print("   üìä ICDF: x such that P(X ‚â§ x) = p - inverse function")
print("   üéØ Transformation: Z = (X - Œº)/œÉ, X = œÉZ + Œº")
print("   üåü These functions are fundamental to statistical inference")

print("üåü" + "="*60 + "üåü")
print("üé™ CDF and ICDF mastery demonstrated! üé™")
```
