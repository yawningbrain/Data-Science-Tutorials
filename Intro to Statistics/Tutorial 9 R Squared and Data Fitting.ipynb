```python
# üé® Welcome to the Data Fitting Art Gallery! Let's set up our creative tools! üé®
print("üé® Welcome to the Data Fitting Art Gallery!")
print("üñåÔ∏è Setting up our artistic toolkit...")

# üéØ Import our essential libraries
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import warnings
warnings.filterwarnings('ignore')

# üé® Set up our beautiful plotting style
plt.style.use('default')
sns.set_palette("husl")

# üé™ Create our custom color palette for data fitting artistry
art_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', 
              '#DDA0DD', '#98D8C8', '#F7DC6F', '#BB8FCE', '#85C1E9',
              '#F8BBD9', '#A8E6CF', '#FFD3A5', '#FD6C9E', '#C7CEEA']

print("üåü Artistic tools loaded successfully!")
print("üé® Beautiful visualizations ready!")
print("üöÄ Ready to create data fitting masterpieces!")

# üé® Load the mysterious artistic dataset
print("\nüé® Loading the mysterious dataset...")
print("üéØ Using a clean, educational dataset for optimal learning experience...")
# Create a clean, educational dataset for demonstration
np.random.seed(42)
x = np.linspace(0, 4*np.pi, 200)
true_function = np.sin(1.5 * x)
noise = np.random.normal(0, 0.75, len(x))
my_signal = true_function + noise
data_import = [x, my_signal]
print("‚úÖ Clean educational dataset created successfully!")
print(f"üìä Dataset contains {len(x)} data points")
print(f"üéØ True function: sin(1.5x) with Gaussian noise N(0, 0.75)")

# üé® Extract our artistic data
x = data_import[0]
my_signal = data_import[1]

print(f"\nüé® Dataset Overview:")
print(f"   üìä X values range: {x.min():.2f} to {x.max():.2f}")
print(f"   üìä Y values range: {my_signal.min():.2f} to {my_signal.max():.2f}")
print(f"   üìà Data points: {len(x)}")

print("\nüåü Dataset loaded! Ready to create artistic masterpieces! üåü")
```

```python
# üé® Visualizing Our Artistic Canvas: The Raw Dataset! üé®
print("üé® Creating our artistic canvas...")
print("üìä Let's visualize the mysterious dataset!")

# üé® Create a beautiful visualization of our raw data
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('üé® The Data Fitting Art Gallery: Our Mysterious Canvas üé®', 
             fontsize=18, fontweight='bold', y=0.98)

# Plot 1: Raw data with artistic styling
axes[0].scatter(x, my_signal, c=art_colors[0], alpha=0.7, s=50, edgecolors='white', linewidth=1)
axes[0].set_xlabel('X Values', fontsize=14, fontweight='bold')
axes[0].set_ylabel('f(x) Values', fontsize=14, fontweight='bold')
axes[0].set_title('üìä Raw Mysterious Dataset üìä', fontsize=16, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].set_facecolor('#f8f9fa')

# Add artistic annotations
axes[0].text(0.02, 0.98, f'Data Points: {len(x)}\nRange X: [{x.min():.1f}, {x.max():.1f}]\nRange Y: [{my_signal.min():.1f}, {my_signal.max():.1f}]', 
             transform=axes[0].transAxes, fontsize=11, verticalalignment='top',
             bbox=dict(boxstyle="round,pad=0.3", facecolor=art_colors[1], alpha=0.7))

# Plot 2: Data distribution analysis
axes[1].hist(my_signal, bins=20, color=art_colors[2], alpha=0.7, edgecolor='white', linewidth=1)
axes[1].set_xlabel('f(x) Values', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Frequency', fontsize=14, fontweight='bold')
axes[1].set_title('üìà Distribution of Our Mysterious Data üìà', fontsize=16, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].set_facecolor('#f8f9fa')

# Add statistical annotations
mean_val = np.mean(my_signal)
std_val = np.std(my_signal)
axes[1].axvline(mean_val, color=art_colors[3], linestyle='--', linewidth=3, label=f'Mean: {mean_val:.2f}')
axes[1].axvline(mean_val + std_val, color=art_colors[4], linestyle=':', linewidth=2, label=f'¬±1œÉ: {std_val:.2f}')
axes[1].axvline(mean_val - std_val, color=art_colors[4], linestyle=':', linewidth=2)
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()

# üéØ Data analysis summary
print(f"\nüéØ MYSTERIOUS DATASET ANALYSIS:")
print("="*60)
print(f"üìä Total Data Points: {len(x)}")
print(f"üìà X Range: [{x.min():.2f}, {x.max():.2f}]")
print(f"üìà Y Range: [{my_signal.min():.2f}, {my_signal.max():.2f}]")
print(f"üìä Mean Y Value: {mean_val:.3f}")
print(f"üìä Standard Deviation: {std_val:.3f}")
print(f"üé® Data appears to have: {'High' if std_val > 1 else 'Low'} variability")

print(f"\nüåü **Artistic Observations:**")
print(f"   üé® The data shows a clear pattern with some noise")
print(f"   üìä The distribution appears to be centered around {mean_val:.2f}")
print(f"   üéØ There might be an underlying function hidden in this data")
print(f"   üö® The noise level suggests we need to be careful about overfitting")

print(f"\nüé™ **Next Steps:**")
print(f"   üßÆ We'll fit polynomials of different degrees to this data")
print(f"   üìä We'll calculate R¬≤ scores for each model")
print(f"   üéØ We'll determine which model best captures the true pattern")
print(f"   üö® We'll watch out for the overfitting trap!")

print("\nüåü Canvas ready! Time to create our polynomial masterpieces! üåü")
```

```python
# üé® Creating Polynomial Masterpieces: The Art of Data Fitting! üé®
print("üé® Creating our polynomial masterpieces...")
print("üñåÔ∏è Let's fit polynomials of different degrees to our mysterious data!")

# üéØ Define our artistic polynomial function
def f_hat(x, w):
    """
    üé® Artistic Polynomial Function Creator
    
    Parameters:
    -----------
    x : array-like
        Input values
    w : array-like
        Polynomial coefficients
        
    Returns:
    --------
    array-like
        Polynomial function values
    """
    # Use numpy's polyval for reliable polynomial evaluation
    return np.polyval(w, x)

# üé™ Define our polynomial degrees for artistic experimentation
poly_order = [1, 3, 5, 10, 25]
print(f"üéØ Polynomial Degrees to Test: {poly_order}")

# üé® Create our artistic masterpieces
print(f"\nüé® Creating Polynomial Masterpieces...")
print("="*60)

w = []  # Store coefficients
y_hat = []  # Store fitted values
r = []  # Store R¬≤ values

for i, dim in enumerate(poly_order):
    print(f"üé® Creating Degree {dim} Masterpiece...")
    
    # Fit the polynomial
    w_temp = np.polyfit(x, my_signal, dim)
    w.append(w_temp)
    
    # Generate fitted values
    y_hat.append(f_hat(x, w_temp))
    
    # Calculate R¬≤ score
    slope, intercept, r_value, p_value, std_err = linregress(my_signal, y_hat[-1])
    r.append(r_value)
    
    print(f"   üìä R¬≤ Score: {r_value:.4f}")
    print(f"   üéØ Model Quality: {'Excellent' if r_value > 0.9 else 'Good' if r_value > 0.7 else 'Fair' if r_value > 0.5 else 'Poor'}")

# üé® Create a spectacular visualization of all our masterpieces
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('üé® The Data Fitting Art Gallery: Polynomial Masterpieces üé®', 
             fontsize=18, fontweight='bold', y=0.98)

# Plot 1: All polynomial fits on one canvas
axes[0].scatter(x, my_signal, c=art_colors[0], alpha=0.6, s=30, label='Original Data', zorder=5)

for i, (degree, fitted_values, r_score) in enumerate(zip(poly_order, y_hat, r)):
    color_idx = (i + 1) % len(art_colors)  # Use modulo to cycle through colors
    axes[0].plot(x, fitted_values, color=art_colors[color_idx], linewidth=2.5, 
                   label=f'Degree {degree} (R¬≤ = {r_score:.4f})', alpha=0.8)

axes[0].set_xlabel('X Values', fontsize=14, fontweight='bold')
axes[0].set_ylabel('f(x) Values', fontsize=14, fontweight='bold')
axes[0].set_title('üé® All Polynomial Masterpieces üé®', fontsize=16, fontweight='bold')
axes[0].legend(fontsize=10, loc='upper right')
axes[0].grid(True, alpha=0.3)
axes[0].set_facecolor('#f8f9fa')

# Plot 2: R¬≤ scores comparison
degrees_str = [f'Degree {d}' for d in poly_order]
bar_colors = [art_colors[i % len(art_colors)] for i in range(len(poly_order))]
bars = axes[1].bar(degrees_str, r, color=bar_colors, alpha=0.8, edgecolor='white', linewidth=2)
axes[1].set_xlabel('Polynomial Degree', fontsize=14, fontweight='bold')
axes[1].set_ylabel('R¬≤ Score', fontsize=14, fontweight='bold')
axes[1].set_title('üìä R¬≤ Scores: The Art Critics\' Ratings üìä', fontsize=16, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].set_facecolor('#f8f9fa')

# Add value labels on bars
for bar, score in zip(bars, r):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                   f'{score:.4f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# üéØ Summary statistics
best_model_idx = np.argmax(r)
best_degree = poly_order[best_model_idx]
best_r2 = r[best_model_idx]

print(f"\nüéØ POLYNOMIAL MASTERPIECE SUMMARY:")
print("="*60)
print(f"üìä Models Created: {len(poly_order)}")
print(f"üéØ Degrees Tested: {poly_order}")
print(f"üìà R¬≤ Scores: {[f'{score:.4f}' for score in r]}")

print(f"\nüèÜ **Best Performing Model:**")
print(f"   üé® Degree: {best_degree}")
print(f"   üìä R¬≤ Score: {best_r2:.4f}")
print(f"   üéØ Quality Rating: {'Excellent' if best_r2 > 0.9 else 'Good' if best_r2 > 0.7 else 'Fair' if best_r2 > 0.5 else 'Poor'}")

print(f"\nüåü **Artistic Observations:**")
print(f"   üìà R¬≤ scores range from {min(r):.4f} to {max(r):.4f}")
print(f"   üéØ {'Higher degree models generally perform better' if max(r) > min(r) else 'All models perform similarly'}")
print(f"   üö® {'Warning: Very high R¬≤ scores might indicate overfitting!' if max(r) > 0.95 else 'R¬≤ scores appear reasonable'}")

print(f"\nüé™ **The Big Question:**")
print(f"   ü§î Is the model with the highest R¬≤ score truly the best?")
print(f"   üö® Or is it memorizing noise instead of learning the true pattern?")
print(f"   üéØ Only the next challenge will reveal the truth!")

# üéØ Analysis and conclusions
print(f"\nüéØ POLYNOMIAL MASTERPIECE ANALYSIS:")
print("="*60)
print(f"üìä Total Models Created: {len(poly_order)}")
print(f"üéØ Degrees Tested: {poly_order}")
print(f"üìà R¬≤ Scores: {[f'{score:.4f}' for score in r]}")

print(f"\nüèÜ **Best Performing Model:**")
print(f"   üé® Degree: {best_degree}")
print(f"   üìä R¬≤ Score: {best_r2:.4f}")
print(f"   üéØ Quality Rating: {'Excellent' if best_r2 > 0.9 else 'Good' if best_r2 > 0.7 else 'Fair' if best_r2 > 0.5 else 'Poor'}")

print(f"\nüåü **Artistic Observations:**")
print(f"   üìà R¬≤ scores range from {min(r):.4f} to {max(r):.4f}")
print(f"   üéØ {'Higher degree models generally perform better' if max(r) > min(r) else 'All models perform similarly'}")
print(f"   üö® {'Warning: Very high R¬≤ scores might indicate overfitting!' if max(r) > 0.95 else 'R¬≤ scores appear reasonable'}")

print(f"\nüé™ **The Big Question:**")
print(f"   ü§î Is the model with the highest R¬≤ score truly the best?")
print(f"   üö® Or is it memorizing noise instead of learning the true pattern?")
print(f"   üéØ Only the next challenge will reveal the truth!")

print("\nüåü Polynomial masterpieces complete! Ready for the truth reveal! üåü")
```

### üé™ The Shocking Truth:

#### üåü **The Hidden Reality:**

$$f(x) = \sin(1.5x)$$

The data you've been analyzing was created by adding Gaussian noise $\mathcal{N}(0, 0.75)$ to this beautiful sinusoidal function. This is a perfect example of how **real-world data** often contains noise that obscures the true underlying pattern.

#### üö® **The Overfitting Trap:**
Now we'll discover the **overfitting illusion** - when models with the highest R¬≤ scores on training data actually perform **worse** on the true underlying function. This is one of the most important lessons in machine learning and data science!

### üé™ Your Mission:

```python
# üåü The Truth Revealed: Exposing the Overfitting Illusion! üåü
print("üåü The moment of truth has arrived!")
print("üéØ Let's reveal the true function and expose the overfitting trap!")

# üåü Define the true underlying function
def funct(x):
    """
    üåü The True Underlying Function
    
    Parameters:
    -----------
    x : array-like
        Input values
        
    Returns:
    --------
    array-like
        True function values: sin(1.5x)
    """
    signal = np.sin(1.5 * x)
    return signal

# üéØ Calculate the true function values
true_function = funct(x)

# üìä Calculate R¬≤ scores between our models and the TRUE function
print(f"\nüìä Calculating True R¬≤ Scores...")
print("="*60)

r_true = []
for i, (degree, fitted_values) in enumerate(zip(poly_order, y_hat)):
    slope, intercept, r_value, p_value, std_err = linregress(true_function, fitted_values)
    r_true.append(r_value)
    print(f"üé® Degree {degree}: True R¬≤ = {r_value:.4f}")

# üé® Create a spectacular truth-revealing visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('üåü The Truth Revealed: Exposing the Overfitting Illusion! üåü', 
             fontsize=18, fontweight='bold', y=0.98)

# Plot 1: True function vs all models
axes[0].plot(x, true_function, color=art_colors[0], linewidth=4, label='True Function: sin(1.5x)', alpha=0.9)
axes[0].scatter(x, my_signal, c=art_colors[1], alpha=0.4, s=20, label='Noisy Data', zorder=5)

for i, (degree, fitted_values, true_r2) in enumerate(zip(poly_order, y_hat, r_true)):
    color_idx = (i + 2) % len(art_colors)  # Use modulo to cycle through colors
    axes[0].plot(x, fitted_values, color=art_colors[color_idx], linewidth=2, 
                   label=f'Degree {degree} (True R¬≤ = {true_r2:.4f})', alpha=0.7)

axes[0].set_xlabel('X Values', fontsize=14, fontweight='bold')
axes[0].set_ylabel('f(x) Values', fontsize=14, fontweight='bold')
axes[0].set_title('üåü True Function vs All Models üåü', fontsize=16, fontweight='bold')
axes[0].legend(fontsize=9, loc='upper right')
axes[0].grid(True, alpha=0.3)
axes[0].set_facecolor('#f8f9fa')

# Plot 2: Training R¬≤ vs True R¬≤ comparison
degrees_str = [f'Degree {d}' for d in poly_order]
x_pos = np.arange(len(degrees_str))
width = 0.35

bars1 = axes[1].bar(x_pos - width/2, r, width, label='Training R¬≤', color=art_colors[2], alpha=0.8)
bars2 = axes[1].bar(x_pos + width/2, r_true, width, label='True R¬≤', color=art_colors[3], alpha=0.8)

axes[1].set_xlabel('Polynomial Degree', fontsize=14, fontweight='bold')
axes[1].set_ylabel('R¬≤ Score', fontsize=14, fontweight='bold')
axes[1].set_title('üìä Training R¬≤ vs True R¬≤: The Overfitting Trap! üìä', fontsize=16, fontweight='bold')
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(degrees_str)
axes[1].legend(fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].set_facecolor('#f8f9fa')

# Add value labels
for bar, score in zip(bars1, r):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                   f'{score:.3f}', ha='center', va='bottom', fontsize=10)
for bar, score in zip(bars2, r_true):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                   f'{score:.3f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# üéØ Overfitting analysis
overfitting_gap = [train - true for train, true in zip(r, r_true)]
best_training_idx = np.argmax(r)
best_true_idx = np.argmax(r_true)
best_training_degree = poly_order[best_training_idx]
best_true_degree = poly_order[best_true_idx]
best_true_r2 = max(r_true)

print(f"\nüåü TRUTH REVEALED: OVERFITTING ANALYSIS")
print("="*70)

print(f"\nüìä **True Function Analysis:**")
print(f"   üåü True Function: f(x) = sin(1.5x)")
print(f"   üìä Noise Level: Gaussian N(0, 0.75)")
print(f"   üéØ Data Points: {len(x)}")

print(f"\nüèÜ **Best Model Identification:**")
print(f"   üìä Training R¬≤ Best: Degree {best_training_degree} (R¬≤ = {r[best_training_idx]:.4f})")
print(f"   üåü True R¬≤ Best: Degree {best_true_degree} (R¬≤ = {best_true_r2:.4f})")
print(f"   üéØ {'Same model is best for both!' if best_training_degree == best_true_degree else 'Different models are best!'}")

print(f"\nüö® **Overfitting Analysis:**")
print(f"   üìà Overfitting Gaps: {[f'{gap:.4f}' for gap in overfitting_gap]}")
print(f"   üö® Most Overfitted: Degree {poly_order[np.argmax(overfitting_gap)]} (Gap = {max(overfitting_gap):.4f})")
print(f"   üåü Least Overfitted: Degree {poly_order[np.argmin(overfitting_gap)]} (Gap = {min(overfitting_gap):.4f})")

print(f"\nüåü **Key Insights:**")
print(f"   üéØ {'Higher degree models show more overfitting' if max(overfitting_gap) > 0.1 else 'Overfitting is minimal across all models'}")
print(f"   üìä Training R¬≤ scores are {'significantly higher' if max(r) - max(r_true) > 0.1 else 'similar to'} true R¬≤ scores")
print(f"   üö® {'Warning: High overfitting detected!' if max(overfitting_gap) > 0.2 else 'Models appear to generalize well'}")

print(f"\nüé™ **The Overfitting Lesson:**")
print(f"   üåü The model with the highest training R¬≤ is NOT always the best choice")
print(f"   üìä True performance on unseen data is what matters most")
print(f"   üö® Validation techniques are essential for model selection")
print(f"   üéØ Balance between bias and variance is crucial")

# üéØ Comprehensive analysis and conclusions
print(f"\nüåü TRUTH REVEALED: COMPREHENSIVE ANALYSIS")
print("="*70)

print(f"\nüìä **True Function Analysis:**")
print(f"   üåü True Function: f(x) = sin(1.5x)")
print(f"   üìä Noise Level: Gaussian N(0, 0.75)")
print(f"   üéØ Data Points: {len(x)}")

print(f"\nüèÜ **Best Model Identification:**")
print(f"   üìä Training R¬≤ Best: Degree {best_training_degree} (R¬≤ = {r[best_training_idx]:.4f})")
print(f"   üåü True R¬≤ Best: Degree {best_true_degree} (R¬≤ = {best_true_r2:.4f})")
print(f"   üéØ {'Same model is best for both!' if best_training_degree == best_true_degree else 'Different models are best!'}")

print(f"\nüö® **Overfitting Analysis:**")
print(f"   üìà Overfitting Gaps: {[f'{gap:.4f}' for gap in overfitting_gap]}")
print(f"   üö® Most Overfitted: Degree {poly_order[np.argmax(overfitting_gap)]} (Gap = {max(overfitting_gap):.4f})")
print(f"   üåü Least Overfitted: Degree {poly_order[np.argmin(overfitting_gap)]} (Gap = {min(overfitting_gap):.4f})")

print(f"\nüåü **Key Insights:**")
print(f"   üéØ {'Higher degree models show more overfitting' if max(overfitting_gap) > 0.1 else 'Overfitting is minimal across all models'}")
print(f"   üìä Training R¬≤ scores are {'significantly higher' if max(r) - max(r_true) > 0.1 else 'similar to'} true R¬≤ scores")
print(f"   üö® {'Warning: High overfitting detected!' if max(overfitting_gap) > 0.2 else 'Models appear to generalize well'}")

print(f"\nüé™ **The Overfitting Lesson:**")
print(f"   üåü The model with the highest training R¬≤ is NOT always the best choice")
print(f"   üìä True performance on unseen data is what matters most")
print(f"   üö® Validation techniques are essential for model selection")
print(f"   üéØ Balance between bias and variance is crucial")

print("\nüåü Truth revealed! The overfitting illusion has been exposed! üåü")
```
